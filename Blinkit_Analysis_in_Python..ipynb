# **Data Cleaning**

```python
print(df['Item Fat Content'].unique())  # unique() prints all unique values present in the Item Fat Content column.
```

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({'LF': 'Low Fat',
                                                         'low fat': 'Low Fat',
                                                         'reg': 'Regular'})    # replace() is used to change specific values in a column or DataFrame with new values.
```

```python
print(df['Item Fat Content'].unique())
```

# **BUSINESS REQUIREMENTS**

### **KPI's REQUIREMENTS**

```python
# Total Sales
total_sales = df['Sales'].sum()
print('Total Sales:',total_sales)

# Average Sales
avg_sales = df['Sales'].mean()
print('Average Sales:',avg_sales)

# No of Item sold
no_of_items_sold = df['Sales'].count()
print('No of Item sold:',no_of_items_sold)

# Average Rating
avg_rating = df['Rating'].mean()
print('Average Rating:',avg_rating)
```

# **CHARTS REQUIREMENTS**

### **1. TOTAL SALES BY FAT CONENT**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()  # Groups the data by Item Fat Content and Calculates the Total Sales for each group.

plt.title('Totale Sales by Fat Content')  # Title is used to set the title of a mathplot lob plot.
plt.pie(sales_by_fat, labels=sales_by_fat.index,
        autopct='%.1f%%',  # Autopct shows percentages.
        startangle=90)     # Startangle rotates the chart.
plt.show()   # Displays the plot on the screen
```

## **This chart shows Sales by Fat Content**

**Low Fat products contribute 65% of total sales**

**Regular products contribute 35% of total sales**

**Conclusion: Low Fat items sold significantly more then Regular items**




### **2. TOTAL SALES BY ITEM TYPE**

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False) # Groups the data by Item Type and Calculates the Total Sales for each group with descending order.

plt.figure(figsize=(10,6))  # It set the figure Size.
bars = plt.bar(sales_by_type.index, sales_by_type.values) # Creating the bar chart,
                                                          # sales_by_type.index---- categories on the X-axis,
                                                          # sales_by_type.values--- corresponding values on the Y-axis.


plt.title('Total Sales by Item Type')  # Title is used to set the title of a mathplot lob plot.
plt.xticks(rotation=90)  # Rotates the x-axis labels by 90 degrees to improve readability.
plt.xlabel('Item Type')  # It set the label for the x-axis of a plot.
plt.ylabel('Total Sales') # It set the label for the y-axis of a plot.

for bar in bars:
  plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
           f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8)  # It adds value labels on top of each bar in a bar chart.

plt.tight_layout() # Automatically adjusts spacing to prevent labels and titles from overlapping.

plt.show()  # Displays the plot on the screen.
```

## **This chart shows Total Sales by Item Type**

**Fruits & Vegetables have the highest sales**

**Snack Foods are the second highest**

**Household and Frozen Foods show strong sales**

**Dairy, Canned, Baking Goods have medium sales**

**Seafood has the lowest sales**

**Sales are higher for daily essential items**

### **3. Fat Content by Outlet for Total Sales**

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content']) ['Sales'].sum().unstack()  # Grouping 
                                                                                              # unstack() Converts grouped data into columns format.

grouped = grouped [['Regular', 'Low Fat']]

ax = grouped.plot(kind='bar', figsize=(8, 5), title='Outlet Tier by Item Fat Content')  # grouped.plot(...), creates a plot from the dataframe grouped.
                                                                                        # kind='bar', makes it a bar chart.
                                                                                        # figsize=(8, 5), sets chart size: 8 inches wide, 5 inches height.
                                                                                        # title='Outlet Tier by Item Fat Content', sets the chart title.
plt.xticks(rotation=30)  # Rotates the x-axis labels by 90 degrees to improve readability.
plt.xlabel('Outlet Location Tier')  # It set the label for the x-axis of a plot.
plt.ylabel('Total Sales') # It set the label for the y-axis of a plot.
plt.tight_layout() # Automatically adjusts spacing to prevent labels and titles from overlapping.
plt.show()  # Displays the plot on the screen.
```

## **This chart shows Fat Content by Outlet for Total Sales**

**Tier 3 has the highest total sales for both Regular and Low Fat items.**

**Low Fat items are selling more than Regular items in all tiers.**

**Sales trend increases like this:
Tier 1 < Tier 2 < Tier 3 (for both fat types)**

**Tier 3 outlets generate the most sales, and Low Fat products have higher sales overall.**

## **4. Total Sales by Outlet Establishment**

```python
sales_by_year = df.groupby('Outlet Establishment Year') ['Sales'].sum().sort_index() # Grouping `
                                                                                     # sort_index(),`sorts data by index order.

plt.figure(figsize=(9,5))  # It set the figure Size.
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-') # x-axis: sales_by_year.index (years)
                                                                               # y-axis: sales_by_year.values (sales values)
                                                                               # marker='o', shows dot points
                                                                               # linestyle='-',  draws a solid line

plt.grid() # `Adds grid lines to the plot (makes values easier to read).
plt.xlabel('Outlet Establishment Year')  # It set the label for the x-axis of a plot.
plt.ylabel('Total Sales')  # It set the label for the y-axis of a plot.
plt.title('Outlet Establishment') # Title is used to set the title of a mathplot lob plot.

for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8) # This loop **adds sales value text labels** on each point in the line chart.

plt.tight_layout() # Automatically adjusts spacing to prevent labels and titles from overlapping.
plt.show()  # Displays the plot on the screen.
```

## **This chart shows Total Sales by Outlet Establishment Year**

**Highest sales** is in **1998 (~204,522).**

**Sales stay almost stable around 129K133K for most years.**

**Lowest sales is in 2011 (~78,132) (big drop).**

**After 2011, sales **recover back to ~130K** and remain steady.**


## **5. Sales by Outlet Size**

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

plt.title('Outlet Size')  # Title is used to set the title of a mathplot lob plot.
plt.pie(sales_by_size, labels=sales_by_size.index,
        autopct='%.1f%%',  # Autopct shows percentages.
        startangle=90)     # Startangle rotates the chart.

plt.tight_layout() # Automatically adjusts spacing to prevent labels and titles from overlapping.
plt.show()   # Displays the plot on the screen
```

## **This pie chart shows Sales by Outlet Size**

**Medium outlets have the highest sales: 42.3%**

**Small outlets are next: 37.0%**
  

**High outlets have the lowest sales: 20.7%**

**Medium size outlets generate most of the total sales.**

## **6. Sales by Outlet Location**

```python
sales_by_location = df.groupby('Outlet Location Type') ['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)           # grouping 

plt.figure(figsize=(8, 3))   # It set the figure Size.
ax = sns.barplot(x='Sales', y='Outlet Location Type', data=sales_by_location)  # Creating the bar chart.

plt.title('Total Sales by Outlet Location Type') # Title is used to set the title of a mathplot lob plot.
plt.xlabel('Total Sales') # It set the label for the x-axis of a plot.
plt.ylabel('Outlet Location Type') # It set the label for the y-axis of a plot.

plt.tight_layout() # Automatically adjusts spacing to prevent labels and titles from overlapping.
plt.show()   # Displays the plot on the screen
```

## **This bar chart shows Total Sales by Outlet Location Type**

**Tier 3 has the highest total sales (~470K).**

**Tier 2 is second highest (~395K).**

**Tier 1 has the lowest sales (~335K).**

**Sales increase from Tier 1 to Tier 3.**
